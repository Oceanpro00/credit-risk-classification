```python
# Import the modules
import numpy as np
import pandas as pd
from pathlib import Path
from sklearn.metrics import confusion_matrix, classification_report, accuracy_score, precision_score, recall_score
from sklearn.preprocessing import StandardScaler
```

---

## Split the Data into Training and Testing Sets

### Step 1: Read the `lending_data.csv` data from the `Resources` folder into a Pandas DataFrame.

```python
# Read the CSV file from the Resources folder into a Pandas DataFrame
lending_data_csv = pd.read_csv(Path("Resources/lending_data.csv"))

# Review the DataFrame
lending_data_csv
```

### Step 2: Create the labels set (`y`)  from the “loan_status” column, and then create the features (`X`) DataFrame from the remaining columns.

```python
# Separate the data into labels and features

# Separate the y variable, the labels
y = lending_data_csv['loan_status']

# Separate the X variable, the features
X = lending_data_csv.drop(columns='loan_status')
```

```python
# Review the y variable Series
display(y.head())
display(y.describe())
```

```python
# Review the X variable DataFrame
display(X.head())
display(X.describe())
```

```python
# Scale the X variable Dataframe Numerical Columns
numerical_columns = X.columns
scaler = StandardScaler()
X_scaled =  scaler.fit_transform(X[numerical_columns])

# Create Dataframe from X scaled Data
X_scaled_df = pd.DataFrame(X_scaled, columns=numerical_columns)

# Review X scaled Dataframe
X_scaled_df
```

### Step 3: Split the data into training and testing datasets by using `train_test_split`.

```python
# Import the train_test_learn module
from sklearn.model_selection import train_test_split

# Split the data using train_test_split
# Assign a random_state of 1 to the function
X_train, X_test, y_train, y_test = train_test_split(X_scaled_df, y, random_state=1)
```

---

## Create a Logistic Regression Model with the Original Data

###  Step 1: Fit a logistic regression model by using the training data (`X_train` and `y_train`).

```python
# Import the LogisticRegression module from SKLearn
from sklearn.linear_model import LogisticRegression

# Instantiate the Logistic Regression model
# Assign a random_state parameter of 1 to the model
logistic_classifier = LogisticRegression(solver="lbfgs", random_state=1)

# Fit the model using training data
logistic_classifier.fit(X_train, y_train)
```

### Step 2: Save the predictions on the testing data labels by using the testing feature data (`X_test`) and the fitted model.

```python
# Make a prediction using the testing data
logistic_classifier_predictions = logistic_classifier.predict(X_test)

# Check Predictions return
display(logistic_classifier_predictions)

# Compare to y_test Data in Dataframe
pd.DataFrame({"Logistic Predictions":logistic_classifier_predictions, "y_test Scores":y_test})
```

### Step 3: Evaluate the model’s performance by doing the following:

* Generate a confusion matrix.

* Print the classification report.

```python
# Generate a confusion matrix for the model
log_matrix = confusion_matrix(y_test, logistic_classifier_predictions)

# display Logistic Regression Confusion Matrix
display(log_matrix)

# Clean up and 'translate' 
print("Logistic Regression model details:")
print(f"- Correctly classified {log_matrix[0,0]} points as 0 (Healthy Loans) but got {log_matrix[0,1]} points Incorrect")
print(f"- Correctly classified {log_matrix[1,1]} points as 1 (High-Risk Loan) but got {log_matrix[1,0]} points Incorrect")
```

```python
# Print the classification report for the model
target_names = ["Low-Risk Applicant (0)", "High-Risk Applicant (1)"]
print(classification_report(y_test, logistic_classifier_predictions, target_names=target_names))
```

```python
# Print Accuracy Score for the model
accuracy = accuracy_score(y_test, logistic_classifier_predictions) * 100
display(f"Model Accuracy: {round(accuracy, 2)}%")

# Print Precision Score for the model
precision = precision_score(y_test, logistic_classifier_predictions) * 100
display(f"Model Precision: {round(precision, 2)}%")

# Print Recall Score for the model
recall = recall_score(y_test, logistic_classifier_predictions) * 100
display(f"Model Recall: {round(recall, 2)}%")
```

**Question:** How well does the logistic regression model predict both the `0` (healthy loan) and `1` (high-risk loan) labels?

**Answer:** The logistic regression model predicts both labels with a high level of accuracy.
